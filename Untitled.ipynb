```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from warnings import filterwarnings
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
filterwarnings('ignore')
```

```python
data = pd.read_csv('datasets/Life Expectancy Data.csv')
```

```python
nonPop = data[data["Population"].isna() == False]
```

```python
nonPop.dropna(inplace = True)
```

```python
nonPop['StatusNum'] = nonPop['Status'].apply(lambda x: 0 if x == "Developing" else 1)
```

```python
nonPop.drop(['Country', 'Year', 'Status'], axis = 1, inplace = True)
```

```python
X = nonPop[['Income composition of resources', 'Schooling', 
             'Diphtheria ', 'GDP', 'Polio', ' BMI ', 'Alcohol', 'percentage expenditure']]
Y = nonPop[['Life expectancy ']]
```

```python
scale = StandardScaler()
scaledDatasetX = scale.fit_transform(X)
scaledDatasetY = scale.fit_transform(Y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaledDatasetX, scaledDatasetY, test_size=0.3, random_state=50)
```

```python
pd.DataFrame(x_train)
```

```python
linearModel = LinearRegression(fit_intercept=True, normalize=True)
model = linearModel.fit(x_train, y_train)
yPred = model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
plt.scatter(y_test, yPred)
```

```python
plot = plt.figure(figsize=(15,10))
sns.regplot(y_test, yPred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, yPred)
mse = mean_squared_error(y_test, yPred)
```
